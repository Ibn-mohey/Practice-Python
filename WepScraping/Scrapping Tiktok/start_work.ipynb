```python
import sqlite3
import re
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time

# Set up SQLite database and create a table
def setup_database():
    conn = sqlite3.connect('Videos.db')
    cursor = conn.cursor()
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS Videos (
            video_id TEXT PRIMARY KEY,
            username TEXT,
            likes INTEGER,
            comments INTEGER,
            shares INTEGER,
            views INTEGER,
            nickname TEXT,
            tags TEXT,
            title TEXT,
            description TEXT
        )
    ''')
    conn.commit()
    conn.close()

# Function to insert or update video data
def insert_or_update_video(data):
    conn = sqlite3.connect('Videos.db')
    cursor = conn.cursor()
    
    # Insert or update video data
    cursor.execute('''
        INSERT INTO Videos (video_id, username, likes, comments, shares, views, nickname, tags, title, description)
        VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
        ON CONFLICT(video_id) DO UPDATE SET
            username=excluded.username,
            likes=excluded.likes,
            comments=excluded.comments,
            shares=excluded.shares,
            views=excluded.views,
            nickname=excluded.nickname,
            tags=excluded.tags,
            title=excluded.title,
            description=excluded.description
    ''', (
        data['video_id'], data['username'], data['likes'], data['comments'], 
        data['shares'], data['views'], data['nickname'], data['tags'], 
        data['title'], data['description']
    ))
    conn.commit()
    conn.close()

# Function to scrape TikTok video data
def scrape_tiktok_video(urls):
    urls = urls.split('\n')
    urls = compact(urls)
    

    # Set up Selenium WebDriver
    chrome_options = Options()
    chrome_options.add_argument("--window-size=1920,1080")
    chrome_options.add_argument("--mute-audio")
    # chrome_options.add_argument("--headless")
    # chrome_options.add_argument("--disable-gpu")
    # chrome_options.add_argument("--no-sandbox")
    # service = Service('path/to/chromedriver')  # Replace with your ChromeDriver path
    driver = webdriver.Chrome(options=chrome_options)
    for url in urls:
        try:
            # Open the URL
            driver.get(url)
            # time.sleep(5)  # Wait for the page to load
            WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'span[class^="css-1q5uu1o-SpanExpandIcon "]'))).click()
            # Extract video data
            likes = driver.find_element(By.CSS_SELECTOR, 'strong[data-e2e^="like-count"]').text
            comments = driver.find_element(By.CSS_SELECTOR, 'strong[data-e2e^="comment-count"]').text
            shares = driver.find_element(By.CSS_SELECTOR, 'strong[data-e2e^="share-count"]').text
            views = driver.find_element(By.CSS_SELECTOR, 'strong[data-e2e^="undefined-count"]').text
            username = driver.find_element(By.CSS_SELECTOR, 'span[data-e2e^="browse-username"]').text
            nickname = driver.find_element(By.CSS_SELECTOR, 'span[data-e2e^="browser-nickname"]').text
            tags = driver.find_element(By.CSS_SELECTOR, 'div[data-e2e^="video-tag"]').text
            title = driver.find_element(By.CSS_SELECTOR, 'div[data-e2e^="v2t-title"]').text
            description = driver.find_element(By.CSS_SELECTOR, 'div[data-e2e^="v2t-desc"]').text

            # Extract video_id from URL using regex
            match = re.search(r'@([a-zA-Z0-9_]+)/video/(\d+)', driver.current_url)
            if match:
                username_from_url = match.group(1)
                video_id = match.group(2)

                # Prepare data for database
                video_data = {
                    'video_id': video_id,
                    'username': username_from_url,
                    'likes': int(likes.replace('K', '000').replace('M', '000000').replace('.', '')) if likes else 0,
                    'comments': int(comments.replace('K', '000').replace('M', '000000').replace('.', '')) if comments else 0,
                    'shares': int(shares.replace('K', '000').replace('M', '000000').replace('.', '')) if shares else 0,
                    'views': int(views.replace('K', '000').replace('M', '000000').replace('.', '')) if views else 0,
                    'nickname': nickname,
                    'tags': tags,
                    'title': title,
                    'description': description,
                }

                # Insert or update the video data in the database
                insert_or_update_video(video_data)
                print(f"Video {video_id} processed successfully.")
            else:
                print("Failed to extract video_id and username from the URL.")
        finally:
            # Close the browser
            driver.quit()

# Main Execution
# if __name__ == "__main__":
#     setup_database()
#     url = "https://www.tiktok.com/@ovhrys1/video/7264294973908995335"  # Replace with the actual TikTok video URL
#     scrape_tiktok_video(url)
```
