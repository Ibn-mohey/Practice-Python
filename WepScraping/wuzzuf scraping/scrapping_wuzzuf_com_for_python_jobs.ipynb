```python
!pip -q install lxml requests beautifulsoup4
import requests 
from bs4 import BeautifulSoup
import csv
from itertools import zip_longest
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:20.0) Gecko/20100101 Firefox/20.0'}
```

```python
def extract_text(list):
    return [i.text for i in list]
def extract_link(list):
    return ["https://wuzzuf.net"+i.find('a').attrs['href'] for i in list]
```

```python
job_title = []
company_name = []
location = []
skills = []
links = []
requirments = []
```

```python
page_number = 0
while True:
    link = f"https://wuzzuf.net/search/jobs/?a=hpb&q=python&start={page_number}"
    src = requests.get(link , headers = headers).content
    soup = BeautifulSoup(src,'lxml')
    job_titles = soup.find_all('h2',{'class':'css-m604qf'})
    job_title += extract_text(job_titles)
    company_name += extract_text(soup.find_all('a',{'class':'css-17s97q8'}))
    location += extract_text(soup.find_all('span',{'class':'css-5wys0k'}))
    skills += extract_text(soup.find_all('div',{'class':'css-y4udm8'}))    
    links += extract_link(job_titles)
    page_number += 1
    showing = soup.find('li',{'class':'css-8neukt'}).text.split(' ')
    print(f'done page {link}')
    if int(showing[-1]) - int(showing[-3]) == 0:
        break
```

```python
requirments = []
i = 0
for link in links:
  requst = requests.get(link, headers= headers)
  src = requst.content
  soup = BeautifulSoup(src,'lxml')
  print(i)
  i += 1
  #reqs
  reqs = soup.find('div',{'class':'css-1t5f0fr'})
  requirments.append(reqs.get_text(' |'))
```

```python
file_lists = [job_title   , company_name  ,location   ,skills   ,links , requirments]
rows_names = ["job_title" , 'company_name','location' , 'skills','links', 'requirments']

exported = zip_longest(*file_lists)
with open("jobs_text.csv", 'w', newline='',encoding='utf-8-sig') as my_file:
    wr = csv.writer(my_file)
    wr.writerow(rows_names)
    wr.writerows(exported)
```
