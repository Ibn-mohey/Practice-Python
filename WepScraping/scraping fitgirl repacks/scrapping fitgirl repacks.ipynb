```python
def scroll_end(driver):
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        # Scroll down to bottom
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # Wait to load page
        time.sleep(2)

        # Calculate new scroll height and compare with last scroll height
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height
```

```python
link = 'https://fitgirl-repacks.site/'
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get(link)
time.sleep(5)
try:
    #click no notification 
    driver.find_element(By.CSS_SELECTOR,value = 'button[id = "smart_push_smio_not_allow"]').click()
except:
    pass
rows = []
df = pd.DataFrame()
#get the last page 
last_page = driver.find_elements(By.CSS_SELECTOR,value = 'a[class = "page-numbers"]')[-1].text
#go throw pages
for page in range(int(200)+2):
    #scroll to end 
    scroll_end(driver)
    #get all articls
    all_articls = driver.find_elements(By.CSS_SELECTOR,value= 'article')
    #go throw articles in page 
    for article in all_articls:
    #start new row
        row = {}
        #get the title 
        row['title'] = article.find_element(By.CSS_SELECTOR, value = 'h1[class = "entry-title"]').text
        #get the date
        row['date'] = article.find_element(By.CSS_SELECTOR, value = 'time').text
      
    # Genres/Tags # Companies # Languages # Original size # Repack Size
    other_data = article.find_element(By.CSS_SELECTOR,value='p').text
    all_info = other_data.split('\n')
    for i in all_info:
        try:
            key_value = i.split(':')
            row[key_value[0]] = key_value[1]
        except:
            pass
    #end of other data 
    
    #links
    #get list
    links_lists = article.find_elements(By.CSS_SELECTOR,value='a[target="_blank"]')
    links = {}
    for link in links_lists:
        row[link.text] = link.get_attribute('href')
    #repack features 
    try:
        row['Repack_Features'] = article.find_elements(By.CSS_SELECTOR,value='ul')[1].text
    
    except:
        pass
    rows.append(row)
#     df = pd.concat([df, pd.DataFrame(rows)])
#     print(len(df))
    print(page)
  #hit next page 
  # next_page = driver.find_element(By.CSS_SELECTOR,value = 'a[class = "next page-numbers"]').click()
    try:
        next_page = driver.find_element(By.CSS_SELECTOR,value = 'a[class = "next page-numbers"]').get_attribute("href")
        driver.get(next_page)
    except:
        print("end")
        break
```
